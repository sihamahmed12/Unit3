### Step 5 - Create Plots for Each Variable in Dataset

```python
# import seaborn

import seaborn as sns
# set standard plot type as seaborn plot
sns.set()

# import matplotlib
import matplotlib.pyplot as plt

# write command for plots to appear in jupyter cells automatically
%matplotlib inline
```

```python
################################### TO DO #########################################
####### FOLLOW PROJECT INSTRUCTIONS TO COMPLETE THIS STEP IN THE CELLS BELOW ######
```

### Plot for 'age' column

```python
# create plot for age column

# sets figure size
plt.figure(figsize=(12,6))

# establishes figure with data and plot type
ax = sns.histplot(data=df, x='age', palette="viridis", bins=8)

# sets x and y labels
plt.xlabel("Age", fontsize=14)
plt.ylabel("Number of Managers", fontsize=14)

# sets title
plt.title("Distribution of Manager Ages in Survey", fontsize=15);
```

##### Insights:

[There is a positive correlation between age and number of managers. As age increases, so does the number of managers. When age is 65 the amount of managers stays steady] 


### Plot for 'industry' column

```python
# create plot for industry column
plt.figure(figsize=(12,6))

ax = sns.countplot(data=df, x='industry',palette="icefire")

plt.xticks(rotation=90)
plt.xlabel("Industry", fontsize=13)
plt.ylabel("Number of Managers in Industry", fontsize=13)
plt.title("Industries in Survey", fontsize=15);
```

##### Insights:

[The number of managers and industry type are roughly all equal]. 

### Plot for 'salary' column

```python
# create plot for salary column
sns.displot(data = df, x = 'salary', kind = 'hist', bins=30)

plt.xlabel("Salary", fontsize=14)
plt.ylabel("Count", fontsize=14)
plt.title("Distribution of Salary Among Managers", fontsize=14);
```

```python
sns.displot(data = df, x = 'salary', kde = True, palette="Greens", edgecolor="black", bins=25, line_kws={"linewidth":2})

plt.xlabel("Salary", fontsize=14)
plt.ylabel("Count", fontsize=14)
plt.title("Distribution of Salary Among Managers", fontsize=16);
```

##### Insights:

[Mean salary for managers is around 100000.]

### What is the correlation between salary and education?

#### Your Hypothesis = [There is a positive correlation between salary and education. ]

[Yes. P-value is 0. ]

### What is the correlation between salary and age?

#### Your Hypothesis [There is a positive correelation between salary and age.]

```python
# calculate correlation and pvalue of salary and age with pearsonr method
spearmanr(df['salary'], df['age'])
```

[Data shows that there is a positive correlation between salary and age.]

#### Are these results statistically significant? How do you know? 

[Yes, results are statistically significant. P-value = 0. ]

#### Crosstab of average salary grouped by education and industry

#### Heatmap of crosstab of average salary grouped by education and industry

```python
# create heatmap with annotated mean salary in each industry and education level

plt.figure(figsize=(16,10))

ax = sns.heatmap(crosstab, annot=True, linewidths=.5, cmap='PiYG', fmt='g', xticklabels=['High School', 'Some College', "Bachelor's Degree", "Master's Degree"])

plt.xlabel("Education Level", fontsize=14)
plt.ylabel("Industry", fontsize=14)

plt.title("Manager Salaries Across Industries by Education Level", fontsize=18)
```

[Data shows that people with Master's have the highest salaries and individuals with high school salaries have the lowest.]
